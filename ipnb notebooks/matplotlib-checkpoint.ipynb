```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,5,1)

#necessary marked as ###
plt.figure(1,figsize=(8,5))              #plotting
plt.plot(x,x,'b-', label='x1') ### 
plt.plot(x,np.square(x),'r--',label='x2')
plt.plot(x,np.cos(x),'g^',label='x3') # o 
plt.xticks(x,['a','b','c','d','e']) #for giving xaxis labell individual
#plt.axes([0,10, -10,10])
plt.xlabel('x') ###
plt.ylabel('y') ###
plt.ylim(-5,25)
plt.title('Plotting')
plt.legend()
#plt.figure(figsize=(20,4))
plt.show() ###


plt.figure(2)
plt.subplot(121)
plt.plot(x,x,'b-', label='fig2 subplot1')
plt.yscale('log') #scale y axis to linear, log, logit 
plt.legend()

plt.subplot(122)
plt.plot(x,np.square(x),'ro', label='fig2 subplot2')
plt.legend()
plt.show()

plt.figure(3)
plt.subplot(211)
plt.plot(x,x,'bs', label='fig3 subplot1')
plt.legend()

plt.subplot(212)
plt.plot(x,np.square(x),'gp', label='fig3 subplot2') #p- pentagon
plt.legend()
plt.show()
```

```python
n=np.array([1,2,3,4,5,10])     #boxplot
plt.boxplot(n)
plt.show()
```

```python
plt.scatter(x,np.sqrt(x))  #scatter
plt.show()
```

```python
label=['a','b','c','d','e']     #barplots
data=[1,5,7,5,-6]
plt.bar(range(len(label)),height=data)
plt.xticks(range(len(label)),label)
plt.show()

plt.bar([1,2,5,6,9],height=data)   #bar(pos no on xaxis,height)
plt.xticks([1,2,5,6,9],label)      #xticks(pos no on xaxis, labelling on those pos nos) 
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D   #3D surface

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X,Y=np.meshgrid(X,Y)
Z=np.sqrt(X**2 + Y**2)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```
