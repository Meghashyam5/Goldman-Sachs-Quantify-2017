```python
import numpy as np
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_csv = pd.read_csv('Pokemon.csv')
```

```python
pokemonDf = pd.DataFrame(pokemon_csv)
```

```python
pokemonDf.keys()
```

```python
del pokemonDf['#'], pokemonDf['Name'], pokemonDf['Type 1'], pokemonDf['Type 2']
```

```python
del pokemonDf['Generation']
```

```python
legends=pokemonDf.pop('Legendary')
```

```python
pokemonDf.keys()
```

```python
from sklearn.cross_validation import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(pokemonDf,legends)
```

```python
xtrain.shape
ytrain.shape
```

```python
ytrain=np.array(ytrain)
features=np.array(xtrain.keys())

for i in range(len(features)):
    for j in range(i+1,len(features),1):
        if i!=j:
            for k in range(len(xtrain)):
                if ytrain[k]==False:
                    plt.plot(xtrain.values[k][i],xtrain.values[k][j],'go')
            for k in range(len(xtrain)):
                if ytrain[k]==True:
                    plt.plot(xtrain.values[k][i],xtrain.values[k][j],'r^')
            plt.xlabel(features[i])
            plt.ylabel(features[j])
            plt.show()
```

```python
#logreg
from sklearn.linear_model import LogisticRegression
for c in np.arange(1,10,1):
    log = LogisticRegression(C=c).fit(xtrain,ytrain)
    print(log.score(xtest,ytest))
log = LogisticRegression(C=5).fit(xtrain,ytrain)
print('final:',log.score(xtest,ytest))
print('final train:',log.score(xtrain,ytrain))
```

```python
#linear SVM
from sklearn.svm import LinearSVC
for c in np.arange(0.1,10,.1):
    svc = LinearSVC(C=c).fit(xtrain,ytrain)
    score=svc.score(xtest,ytest)
    #print(score)
    plt.scatter(c,score)
plt.show()
svc = LinearSVC(C=10).fit(xtrain,ytrain)
score=svc.score(xtest,ytest)
print('final:',score)
```

```python
#non-linear SVM
from sklearn.svm import SVC
score=[]
for c in np.arange(1,10):
    for g in np.arange(0.01,1,0.1):
        svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(xtrain,ytrain)
        score.append(svm.score(xtest,ytest))
        
```

```python
np.max(score)
```

```python
from sklearn.grid_search import GridSearchCV
paramGrid=[ {'C':np.arange(0.1,20,0.1), 'gamma':np.arange(0.01,10,0.1), 'kernel':['rbf']} ]
#paramGrid=[ {'C':np.arange(0.001,0.01,0.001), 'gamma':np.arange(0.0001,.001,0.001), 'kernel':['rbf']} ]

grid=GridSearchCV(estimator=SVC(),param_grid=paramGrid)
```

```python
grid.fit(xtrain,ytrain)
```

```python
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(xtrain,ytrain)
svm.score(xtest,ytest)        
```

```python
score=0
cMax=1
gammaMax=0.05
for c in np.arange(0.1,2,0.1):
    for gamma in np.arange(0.01,2,0.01):
        svm=SVC(kernel='rbf', gamma=gamma, C=c)
        svm.fit(xtrain,ytrain)
        temp=svm.score(xtest,ytest)
        if(temp>score):
            cMax=c
            gammaMax=gamma
            score = temp
print('cMax',cMax)
print('gammaMax',gammaMax)
print('score',score)
```

```python
score=0
cMax=1
gammaMax=0.05
for c in np.arange(0.01,15,0.01):
    for gamma in np.arange(0.001,0.2,0.001):
        svm=SVC(kernel='rbf', gamma=gamma, C=c)
        svm.fit(xtrain,ytrain)
        temp=svm.score(xtest,ytest)
        if(temp>score):
            cMax=c
            gammaMax=gamma
            score = temp
print('cMax',cMax)
print('gammaMax',gammaMax)
print('score',score)
```

```python
from sklearn.ensemble import RandomForestClassifier
ranScore=0
nMax=3
mMax=2
ranFinal=RandomForestClassifier()
for n  in range(3,10,1):
    for m in range(2,7,1):
        ran=RandomForestClassifier(max_features=m,n_estimators=n).fit(xtrain,ytrain)
        temp=ran.score(xtest,ytest)
        if(temp>ranScore):
            ranScore=temp
            nMax=n
            mMax=m
            ranFinal=ran
print('Max n:',nMax)
print('Max m:',mMax)
print('final Score', ranScore)
```

```python
ranFinal.score(xtest,ytest)
```
